```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec 

colors = ['#000000', '#ff0000', '#0073cf', '#008b29', '#800080', '#FFA500', '#FFFF00', '#addd8e']

# Colors: 0 Hitam, 1 Merah, 2 Biru, 3 Hijau, 4 Ungu, 5 Orange, 6 Kuning, 7 Hijau Muda

plt.rcParams.update({'font.size': 19})

data = pd.read_csv('/content/rmsd_backbone.xvg',sep='\s+',header=None)

data = pd.DataFrame(data)

x = data[0]

y = data[1]

data2 = pd.read_csv('/content/rmsd_pro_lyco.xvg',sep='\s+',header=None)

data2 = pd.DataFrame(data2)

x2 = data2[0]

y2 = data2[1]

plt.figure (figsize =(15,10))

plt.plot (x, y, alpha = 2, mec = 'k', linewidth=3, markersize = 7, color = colors[0], label="Protein")
plt.plot (x2, y2, alpha = 2, mec = 'k', linewidth=3, markersize = 7, color = colors[7], label= "Ligand")

plt.ylim(0,1)

plt.tick_params (top = True, right = True, direction = 'in',length=10, width=1) 

plt.ylabel('RMSD Ã…')

plt.xlabel('Time (ns)')

plt.title("RMSD")
plt.legend()

# plt.legend()

plt.savefig('RMSD.pdf')
```
